# Why is Data Important? 

# ડેટા સુ કરવા મહત્વપૂર્ણ છે?

## There is so much data out there:

## ડેટા બધે છે, એનું બૌવ કામ આવે છે

### With data, you have quality in your information.

### ડેટા અપ્ર ઇન્ફોરમેશન ને ક્યુઆલિટી અબે છે 

Data-enabled discoveries are powerful. if you are able to use data in making discoveries, you will be able to make more informed and evidence-based observations.

તમે વધારે સાચું માહિતી થી વધારે જાણી શકો છો. 

### With data, you gain variety. 

### ડેટા વધારે વિવિધતા આપે છે 

There are a lot of methods and approaches to working with information. The data itself provides a lot of variety of information that you can make use of.

માહિતીનો ઉપયોગ કરવા અને કાર્ય કરવાની ઘણી બધી રીતો છે.

### Decision Making becomes easier.

### નિર્ણય લેવાનું સરળ બને છે

તમે વાસ્તવિક વિશ્વનું પ્રદર્શન કરી શકો છો અને યોગ્ય નિર્ણયો લઈ શકો છો.

### Bias

### પૂર્વગ્રહો

Sometimes data might be skewed or might be in favor of certain individuals. For example, in a survey about weather in the US, if you only survey people from California, you will have biased results compared to surveying people from different states. 

કેટલીકવાર માહિતી ખોટી હોય છે અને તે તમારા વિચારો અને નિર્ણયોને ટાળી શકે છે.

# What is Machine Learning and Artificial intelligence? 

# મશીન લર્નિંગ 

**Machine Learning** is **Artificial Intelligence**. **Artificial Intelligence** is **Machine Learning**

These are "fields" or sectors within the Computer Science and Data Science world. ML and AI are built of algorithms and designs that are made to automate the work being done with data and computers. 

These include a varitey of algorithms like supervised vs unsupervised model, from which the computer is learning something given information and data. AI makes use of ML to take in data and information and take actions based on that. 


આ ક્ષેત્રો કામ સ્વચાલિત બનાવવા માટે અને માનવની સહાય વિના રચાયેલ છે. આ કરવા માટે, તમે ઘણાં કોડિંગ અને કમ્પ્યુટરનો ઉપયોગ કરી શકો છો.

આ બધા માટે તમારે ચોક્કસપણે ડેટા ની જરૂર પડશે.

### Sub-topic is Classification/Prediction: 

### આગાહી:

For example, you recieve emails that are **spam** (bad) and **ham** (good). 

The **problem** is that you don't know which emails are good and bad. 

In order to **predict** or **classify** emails into "good" and "bad" you can build a Machine Learning model that can make these groups for you. 



ઉદાહરણ તરીકે તમને એક ઇમેઇલ મળે છે જે સારું કે ખરાબ છે. અમે આ ઇમેઇલને સ્પામ કહીએ છીએ
